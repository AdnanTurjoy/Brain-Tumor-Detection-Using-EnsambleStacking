**Ensemble Technique** 





```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sklearn import svm
from sklearn.svm import SVC
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn 
```

```python
df=pd.read_csv("Brain.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df.drop(['output'],axis=1)
x
```

```python
y=df['output']
y
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=40)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
                       max_depth=None, max_features='auto', max_leaf_nodes=None,
                       min_impurity_decrease=0.0, 
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=42,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False)
RF.fit(x_train,y_train)
```

```python
RF.score(x_test,y_test)
```

```python
pred=RF.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
knn.score(x_test,y_test)
```

```python
pred=knn.predict(x_test)
```

**Decision Tree**

```python
dtclf=DecisionTreeClassifier()
rfc = RandomForestClassifier()
knn =  KNeighborsClassifier()
svc = SVC()
```

```python
clf = [dtclf,rfc,knn,svc]
for algo in clf:
    score = cross_val_score( algo,x,y,cv = 5,scoring = 'accuracy')
    print("The accuracy score of {} is:".format(algo),score.mean()*100)
```

**Stacked Model**

```python
dtclf=DecisionTreeClassifier()
rfc = RandomForestClassifier()
knn =  KNeighborsClassifier()
svc = SVC()
clf = [('dtclf',dtclf),('rfc',rfc),('knn',knn),('svc',svc)]
```

```python
lr = LogisticRegression()
stack_model = StackingClassifier( estimators = clf,final_estimator = lr)
stack_model.fit(x_train,y_train)
```

```python
stack_model.score(x_test,y_test)
```

```python
pred=stack_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
